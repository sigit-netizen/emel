#### read dataset

```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import os
```

```python
df = pd.read_csv("dataset\WineQT.csv")
print("Dataset Loaded!")
df.head()
```

```python
df.info()
```

```python
df["quality"].value_counts()
```

#### var inden dan depen

```python
x = df.iloc[:, :-2]
y = df.iloc[:, -2]

y.head()
```

```python
x
```

#### split data test dan train

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=0 , shuffle=False)
y_test
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

#### modeling

```python
from sklearn import tree 

model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
model.fit(x_train, y_train)

print("Training selesai!")
```

#### evaluasi

```python
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = model.predict(x_test)

print("\n=== EVALUASI MODEL NAIVE BAYES ===")
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt


fig =plt.figure(figsize=(150, 60))
tree.plot_tree(model, feature_names=x_train.columns, class_names=True, filled=True, rounded=True, fontsize=12)
plt.title("Decision Tree (CART, default)")
plt.show()
```

#### simpan pohon ke pdf

```python
fig.savefig('decision_tree.pdf', format='pdf',bbox_inches='tight') 
```

#### simpan model

```python

with open('models/id3.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

print("model dan scaler berhasil disimpan")
```
